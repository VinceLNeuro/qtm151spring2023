(a) Create two normal random variables ...


- Call them "vec_x" and "vec_z"
- To generate each variable, use loc = 2, scale = 5, size = 1000
- Plot a histogram of each variable
    - Label the axes and title
    - Each graph should have a different title
 
To get full points, plot the histograms on a grid (See Lecture 6)

```python
# Create two normal random variables
vec_x = np.random.normal(loc=2, scale=5, size=1000)
vec_z = np.random.normal(loc=2, scale=5, size=1000)

# Create a list to store the vectors
ls_vec = [vec_x, vec_z]
name_vec = ['vec_x', 'vec_z']
# Create the plotting area
fig, subfig = plt.subplots(1,2, figsize = (12,6))

index = 1
for i in range(len(ls_vec)):
    subfig[i].hist(ls_vec[i])
    subfig[i].set_xlabel('Value', fontsize = 15, labelpad = 10)
    subfig[i].set_ylabel('Frequency', fontsize = 15, labelpad=10)
    subfig[i].set_title('Figure '+str(index)+'. The histogram showing \n the distribution of '+name_vec[i], fontsize = 15, y = 1.05, fontweight = 'bold')
    # y = 1.05 => increase the height of the title by 0.05 comparing with the default (y=1)
    plt.tick_params(axis = 'both', labelsize = 10)
    index = index + 1

fig.suptitle('Two histograms for normal random variables', fontsize = 20)
fig.tight_layout()

# Final adjustments for the white space between two subplots
plt.subplots_adjust(wspace=0.2)
```
